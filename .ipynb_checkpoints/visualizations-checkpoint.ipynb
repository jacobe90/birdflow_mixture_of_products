```python
def get_losses(fname):
    with open(f"/work/pi_drsheldon_umass_edu/birdflow_modeling/jacob_independent_study/mixture_of_products/experiments/training_variability_10_components/{fname}", "rb") as f:
        losses = pickle.load(f)
    return losses['total']
loss_files = list(filter(lambda fname: "losses" in fname, list(os.listdir("/work/pi_drsheldon_umass_edu/birdflow_modeling/jacob_independent_study/mixture_of_products/experiments/training_variability_10_components"))))
all_losses = np.asarray([get_losses(fname) for fname in loss_files])
```

```python
def ci(y):
    return 1.96 * y.std(axis=0) / np.sqrt(1)
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
losses_mean = all_losses.mean(axis=0)
losses_ci = ci(all_losses)
#print(losses_ci)
ax.plot(np.arange(600), all_losses.mean(axis=0), label="mean mop loss (batch size 40)")
ax.fill_between(np.arange(600), losses_mean - losses_ci, losses_mean + losses_ci, color='red', 
                 alpha=0.15, label="95% confidence interval")
ax.plot(markov_losses, label="markov chain", color="purple")
ax.legend(loc="upper right")
ax.set(
    xlabel=f"training steps",
    ylabel=f"loss value",
    title="Mixture of products training variability (ew=1e-4, dw=1e-2, dp=0.4, n=10)"
    
)
plt.show()
```
